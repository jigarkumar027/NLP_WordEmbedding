```python
from tensorflow.keras.preprocessing.text import one_hot 
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
s = ['you are good', 'you are not good','there is a cow','cow gives us milk','i love apple cake','i love banana cake']
```

```python
voc_size = 10000
```

```python
onehot = [one_hot(word,voc_size)for word in s]
```

```python
onehot
```

```python
SL = 7
embedded = pad_sequences(onehot,padding='pre',maxlen = SL)
```

```python
embedded
```

```python
model = Sequential()
model.add(Embedding(voc_size,10,input_length = 7))
model.compile('adam','mse') 
```

```python
model.summary()
```

```python
a= model.predict(embedded)
```
